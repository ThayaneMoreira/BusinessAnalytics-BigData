```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

#Upload dos dados
cadastro_df = pd.read_csv('cadastro - cadastro.csv', encoding='utf-8')
precos_df = pd.read_csv('preco da gasolina - preco da gasolina.csv', encoding='utf-8')
salario_df = pd.read_csv('salario_profissoes - salario_profissoes.csv', encoding='utf-8')

#Apresentando apenas 3 linhas do dataframe
cadastro_df.head(3)
```

```python
#Apresentando apenas 3 linhas do dataframe
precos_df.head(3)
```

```python
#Apresentando apenas 3 linhas do dataframe
salario_df.head(3)
```

```python
#Filtrando apenas nascidos antes de 1985
cadastro_df[cadastro_df['ano_nascimento'] < 1985].sort_values(by = 'ano_nascimento')
```

```python
#Filtrando apenas mulheres com estado civil Solteira
cadastro_df[(cadastro_df['sexo'] == 'f') & (cadastro_df['estado_civil'] == 'Solteira')]
```

```python
#Incluíndo dados em uma nova coluna
km_rodados = [34,300,150,345,34,23,122,234,454,55,99,100,130,43,32,
          31,320,152,35,14,223,122,244,424,51,91,10,13,4,2]
```

```python
#Acrescentando nova coluna ao dataframe
cadastro_df['km_rodados'] = km_rodados
```

```python
cadastro_df.head(3)
```

```python
#Verificação do tipo do dado
precos_df.dtypes
```

```python
#Alterando o tipo do dado
precos_df['preço'].str.replace(',', '.').astype(float)
```

```python
#Subindo alteração
precos_df['preço'] = precos_df['preço'].str.replace(',', '.').astype(float)
```

```python
precos_df
```

```python
#Unido dataframes
cadastro_compl = pd.merge(cadastro_df, precos_df, on='cidade')
```

```python
#Criando campo calculado ( km * preço)
cadastro_compl['vl_calculado'] = cadastro_compl['km_rodados'] * cadastro_compl['preço']
```

```python
#Arredondando valores
round(cadastro_compl['quanta_vezes_na_semana_pratica_esporte'].mean(),2)
```

```python
#identificando valores nulos
cadastro_compl.isnull().sum()
```

```python
#Preenchendo campos nulos com uma média
cadastro_compl['quanta_vezes_na_semana_pratica_esporte'].fillna(2)
```
