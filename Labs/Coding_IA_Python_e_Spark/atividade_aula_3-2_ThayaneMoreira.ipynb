## Carregando os dados

```python
import requests
import zipfile
from scipy.io import arff
import pandas as pd
import io


url = "https://dados-lab-python.s3.amazonaws.com/carro_atividade2.csv"
r = requests.get(url, stream=True)

df_carros = pd.read_csv(url,',',names=['preço_compra','preço_manutenção','portas','pessoas','porta-malas','segurança','classe'])
```

```python
df_carros
```

```python
class_mapper = {'n_aceito':0,'aceito':1,'bom':2,'m_bom':3}
df_carros['classe'] = df_carros['classe'].replace(class_mapper)
```

```python
df_carros.head(3)
```

## 2. Codifique  os componentes (features) ordinais.

```python
# Escreva seu código aqui
tamanho_valor = {'pequeno': 1, 'med': 2,'grande':3 }
print(tamanho_valor)
```

```python
df_carros['porta-malas-tamanho'] = df_carros['porta-malas'].replace(tamanho_valor)
df_carros
```

## 3.  Mova a coluna alvo (target) para a primeira posição

```python
# Escreva seu código aqui
#df_carros.columns
cols = df_carros.columns.tolist()
cols = cols[-1:] + cols[:-1] # Concatena o ultimo elemento da lista no inicio da mesma.
df_carros = df_carros[cols]

df_carros.columns
```

## 4. Divida os dados em três conjuntos: Criação do modelo, Validação e Teste

```python
# Escreva seu código aqui
from sklearn.model_selection import train_test_split
train, test_and_validate = train_test_split(df_carros, test_size = 0.2, random_state=42, stratify = df_carros['classe'])

test, validate = train_test_split(test_and_validate, test_size=0.5, random_state=42, stratify = test_and_validate['classe'])

print(train.shape)
print(test.shape)
print(validate.shape)
```

```python
print(train['classe'].value_counts())
print(test['classe'].value_counts())
print(validate['classe'].value_counts())
```
